## Chapter 10 Transformations

##### Split/Apply/Combine

```python
# find top net feelies
def find_max(df, n=5, column="NET FEELIES"):
    return df.sort_values(by=column)[-n:]
```

```python
# find top feelies according to item
cleaned.groupby("NET FEELIES").apply(find_max)
```
