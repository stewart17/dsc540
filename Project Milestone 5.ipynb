```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sqlite3
import sqlalchemy
```

```python
# import all cleaned data sets
vaccines = pd.read_csv("vaccines_cleaned.csv")
energy = pd.read_csv("energy_cleaned.csv")
cases = pd.read_csv("cases_cleaned.csv")
```

```python
# create and open database
conn = sqlite3.connect("project.db")
```

```python
c = conn.cursor()
```

```python
c.execute("CREATE TABLE IF NOT EXISTS vaccines (location text, vaccinated number, percent number)")
```

```python
conn.commit()
```

```python
vaccines.to_sql("vaccines", conn, if_exists="replace", index=False)
```

```python
c.execute("CREATE TABLE IF NOT EXISTS covid_data (date text, country text, new_confirmed number, total_confirmed number, new_deaths number, total_deaths number, new_recovered number, total_recovered number)")
```

```python
conn.commit()
```

```python
cases.to_sql("covid_data", conn, if_exists="replace", index=False)
```

```python
c.execute("CREATE TABLE IF NOT EXISTS pollution_data (country text, energy_type text, year text, energy_consumption number, energy_production number, GDP number, population number, energy_intensity_per_capita number, energy_intensity_GDP number, co2_emission number)")
```

```python
conn.commit()
```

```python
energy.to_sql("pollution_data", conn, if_exists="replace", index=False)
```
