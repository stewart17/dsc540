# Project Milestone 4

```python
import pandas as pd
import json
import requests
```

## Loading Dataset

```python
# get summary of cases
response = requests.get('https://api.covid19api.com/summary')
```

```python
# turn into json data
summary = response.json()
```

```python
summary
```

```python
# split data into global vs countries
summary_global = summary["Global"]
summary_countries = summary["Countries"]
```

```python
cases_df = pd.json_normalize(summary_countries)
cases_df
```

```python
# normalize global cases
cases_global = pd.json_normalize(summary_global)
```

```python
# add two back together for starting data set
cases_df = cases_global.append(cases_df)
```

## Transformation 1: Remove Unecessary Columns

The ID, Slug, and CountryCode columns are not necessary to our project so they will be removed.

```python
# drop id, countrycode, slug columns
cases_df = cases_df.drop(["ID", "CountryCode", "Slug"], axis = 1)
```

```python
cases_df
```

## Transformation 2: Replace Headers

To keep the same convention we have in our other datasets, we will rename the columns to have underscores between words.

```python
# rename columns
cases_df = cases_df.rename(columns={"NewConfirmed":"New_confirmed", "TotalConfirmed":"Total_confirmed", "NewDeaths":"New_deaths","TotalDeaths":"Total_deaths", "NewRecovered":"New_recovered", "TotalRecovered":"Total_recovered"})
```

```python
cases_df
```

## Transformation 3: Fix World Row

Because of how the data was formatted when pulled from the API, we had to attatch the world data separately from the rest, and now it isn't named as global in its column. We will fix that.

```python
# fill na in country with World
cases_df["Country"].fillna("World", inplace=True)
```

```python
cases_df
```

## Transformation 4: Reset Index Rows

Concatenating the two data frames has caused the index to be off. So we will reset the index.

```python
# reset index rows
cases_df.reset_index(inplace=True)
```

```python
# drop reindex row
cases_df.drop("index", axis = 1, inplace=True)
```

```python
cases_df
```

## Transformation 5: Reorder the Columns

The data frame is not very readable with the country and date columns all to the left. To make it more readable, we will move those to the left and shift everything to the right.

```python
# reorder rows
cases_df = cases_df.iloc[:,[6,7,0,1,2,3,4,5]]
```
